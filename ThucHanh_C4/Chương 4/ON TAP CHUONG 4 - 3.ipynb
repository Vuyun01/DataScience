# Bài tập 27: Làm việc với Matplotlib
Sử dụng kết quả thống kê được ở Bài tập 20 về nhiệt độ max, min, mean của 6 thành phố:

Yêu cầu:

    1. Vẽ biểu đồ dạng đường biểu diễn nhiệt độ Max của 6 thành phố
    2. Vẽ biểu đồ dạng cột biểu diễn nhiệt độ Mean của 6 thành phố
    3. Vẽ biểu đồ dạng điểm biểu diễn nhiệt độ Min của 6 thành phố

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dict_name = {0:'Hà Nội',
            1:'Vinh',
            2:'Đà Nẵng',
            3:'Nha Trang',
            4:'Tp.HCM',
            5:'Cà Mau',
            6:'Chung'}
index_list = {0:'Max',
             1:'Min',
             2:'Mean'}

data_thongke = pd.read_csv('Thongke.txt',sep ='\t', header = None)
data_thongke_new = data_thongke.rename(columns = dict_name, index = index_list)
data_thongke_new1 = data_thongke_new.iloc[:,:6]
data_thongke_new1
```

```python
#1: Vẽ biểu đồ dạng đường biểu diễn nhiệt độ Max của 6 thành phố

plt.figure(figsize= (10,6))
plt.plot(data_thongke_new1.columns,
        data_thongke_new1.iloc[0,:],
        'r*--',
        label = 'ToC - Max',
        markersize = 12)
plt.legend()
plt.ylabel('Nhiệt độ (oC)')
plt.title('Biểu đồ thể hiện nhiệt độ Max của 6 thành phố')
plt.grid(True)
plt.show()
```

```python
#2: Vẽ biểu đồ dạng cột biểu diễn nhiệt độ Mean của 6 thành phố

# data_thongke_new1
plt.rcParams["figure.figsize"] = (10,6)
plt.bar(data_thongke_new1.columns,
        data_thongke_new1.iloc[2,:],
       color = 'black')
plt.ylabel('Nhiệt độ (oC)')
plt.title('Biểu đồ thể hiện nhiệt độ Mean của 6 thành phố')
plt.grid(True)
plt.show()
```

```python
#3: Vẽ biểu đồ dạng điểm biểu diễn nhiệt độ Min của 6 thành phố

# plt.rcParams["figure.figsize"] = (10,6)
plt.rcParams['figure.figsize'] = (10,5)
plt.scatter(data_thongke_new1.columns,
           data_thongke_new1.iloc[1,:],
           marker = 'o',
           s = 300,
           color = 'b',
           alpha = 0.3)
plt.title('Biểu đồ thể hiện nhiệt độ Min của 6 thành phố')
plt.ylabel('Nhiệt độ (oC)')
plt.grid(True)
plt.show()
```
