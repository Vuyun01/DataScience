# Bài tập 26:
1. Đọc dữ liệu từ file Data_Patient.csv
2. Cho biết các cột chứa giá trị khuyết thiếu, xác định các vị trí thiếu dữ liệu và đề xuất phương án xử lý giá trị thiếu.
3. Phát hiện giá trị ngoại lai và xử lý trong dữ liệu (nếu có)

```python
#1:
df_new_patient = pd.read_csv('Data_Excercise\Data_Patient.csv')
df_new_patient
```

```python
#rename columns
dict_title = {'feature_1':'Age',
            'feature_2':'Gender',
            'feature_3':'Type',
            'feature_4':'Blood_pressure',
            'feature_5':'Cholesterol',
            'feature_6':'Heartbeat',
            'feature_7':'Thalassemia',
            'feature_8':'Result'}

df_new_patient.rename(columns = dict_title, inplace=True)
df_new_patient.head()
```

```python
#2:
def missing_values(df):
    missing_value = df.isnull().sum()
    missing_value_pt = 100 * df.isnull().sum()/ len(df)
    missing_table = pd.concat([missing_value,missing_value_pt], axis = 1)
    missing_table_new = missing_table.rename(columns = {0:'So cot missing',1:'Ti le % missing'})
    missing_table_new = missing_table_new[missing_table_new.iloc[:,0] != 0].sort_values('So cot missing', ascending = False).round(2)
    print('Tong so cot trong bang: ',df.shape[1],'\nCac cot bi missing values: ',missing_table_new.shape[0])
    return missing_table_new
missing_values(df_new_patient)
```

```python
df_new_patient[df_new_patient.isnull().any(axis = 1)]
```

```python
#processing missing value
df_new_patient1 = df_new_patient.interpolate(method='linear',limit_direction ='forward').head(10)
df_new_patient1

```

```python
#checking outlier value
plt.figure(figsize=(8,5))
df_new_patient1.boxplot()
df_new_patient1.loc[:,['Cholesterol','Heartbeat']]
```

```python
#xu ly outlier value ve NaN

df_new_patient1.loc[7,'Cholesterol'] = np.NaN
df_new_patient1.loc[1,'Heartbeat'] = np.NaN

#ve bieu do
plt.figure(figsize =(8,5))
df_new_patient1['Cholesterol'].plot(style='-o',color='lightgreen')
df_new_patient1['Heartbeat'].plot(style='--s',color='lightblue')
plt.grid(True)

df_new_patient1.loc[:,['Cholesterol','Heartbeat']]
```

```python
#copy df
df_new_copy = df_new_patient1.copy()
```

```python
#xu ly du lieu NaN
df_new = df_new_copy.interpolate(method='linear', limit_direction='forward')

#ve bieu do
df_new['Cholesterol'].plot(style='-o', color='lightgreen')
df_new['Heartbeat'].plot(style='--s',color='lightblue')
plt.grid(True)

```
