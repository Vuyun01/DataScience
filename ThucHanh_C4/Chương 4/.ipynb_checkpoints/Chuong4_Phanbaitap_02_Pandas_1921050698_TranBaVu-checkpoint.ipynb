## Bài tập 26:
---
* Đọc dữ liệu từ file Data_Patient.csv
* Cho biết các cột chứa giá trị khuyết thiếu, xác định các vị trí thiếu dữ liệu và đề xuất phương án xử lý giá trị thiếu.
* Phát hiện giá trị ngoại lai và xử lý trong dữ liệu (nếu có)

```python
#{HỌC VIÊN VIẾT CODE XỬ LÝ Ở ĐÂY}
import pandas as pd
import numpy as np
df_patients = pd.read_csv('Data_Excercise\Data_Patient.csv')

dict_patients = {'feature_1':'Age',
                'feature_2':'Gender',
                'feature_3':'Type',
                'feature_4':'Blood_pressure',
                'feature_5':'Cholesterol',
                'feature_6':'Heartbeat',
                'feature_7':'Thalassemia',
                'feature_8':'Result'}

df_patients.rename(columns =  dict_patients, inplace = True)
df_patients.info()



```

```python
def values_detect(df):
    missing_values = df.isnull().sum()
    missing_values_pt = 100 * df.isnull().sum() / len(df)
    missing_values_cols = pd.concat([missing_values,missing_values_pt], axis = 1) 
    missing_values_table = missing_values_cols.rename(columns = {0:'Số giá trị missing',1:'Tỉ lệ % missing'})
    missing_values_table = missing_values_table[
        missing_values_table.iloc[:,0] != 0].sort_values('Số giá trị missing', ascending = False).round(2)
    print(f'File dữ liệu bao gồm có: {df.shape[1]} cột \nCó {missing_values_table.shape[0]} cột chứa missing values')
    return missing_values_table
values_detect(df_patients)
```
